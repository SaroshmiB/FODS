```python
def fillgrade(s):
    if(s>=90):
        return 'A'
    elif(s<90 and s>=80):
        return 'B'
    elif(s<80 and s>=70):
        return 'C'
    else:
        return 'D'
df['Grade']=df['Score'].apply(fillgrade)
```
